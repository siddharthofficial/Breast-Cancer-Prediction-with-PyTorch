```python
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2)
```

**Preprocessing**

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
#As output has alphabets instead of 0 and 1
encoder=LabelEncoder()
y_train=encoder.fit_transform(y_train)
y_test=encoder.transform(y_test)
```

**Converting numpy array to tensors**

```python
x_train_tensor=torch.from_numpy(x_train)
y_train_tensor=torch.from_numpy(y_train)
x_test_tensor=torch.from_numpy(x_test)
y_test_tensor=torch.from_numpy(y_test)
```

```python
# Creating a constructor
class mysimpleNN:
  def __init__(self,x):
    # initialing weights with random values and weight tensor should have no. of rows as the no. of input 
    # columns and 1 columns...for each weight there is a random value
    self.weights=torch.rand(x.shape[1],1,dtype=torch.float64,requires_grad=True)
    # As we are only using one perceptron , we need 1 bias only and that is initiliase to 0
    self.bias=torch.zeros(1,dtype=torch.float64,requires_grad=True)

    #calculating forward propogation
  def forward(self,x):
      z=torch.matmul(x,self.weights)+self.bias
      y_pred=torch.sigmoid(z)
      return y_pred
  #calculating loss
  def calculate_loss(self,y_pred,y):
      # as we are using loss , so we want to avoid the situation where we get log 0 or log 1 ,so we clamp our 
      # predicted values  between 10^-7 and 0.9999999
      epsilon=1e-7
      y_pred=torch.clamp(y_pred,epsilon,1-epsilon)

      #using binary cross entropy formula
      loss= -(y*torch.log(y_pred)+(1-y)*torch.log(1-y_pred)).mean()
      return loss
```

```python
learning_rate=0.1
epochs=30
```

```python
model=mysimpleNN(x_train_tensor)
model.bias
```

```python
for epoch in range(epochs):
    #forward pass
    y_pred=model.forward(x_train_tensor)

    #loss 
    loss=model.calculate_loss(y_pred,y_train_tensor)
  
    #backpropagation and using auto grad to calculate gradients
    loss.backward()

    # updating weights
    # as weights are being tracked by autograd , we dont want to calculate gradient while upgrading weights so 
    # for this , we turn off gradient searching using torch.no_grad_
    with torch.no_grad():
     model.weights-=learning_rate*model.weights.grad
     model.bias-=learning_rate*model.bias.grad

    #intialzing those grads to 0 again to prevent multiplication of old gradient to new gradient giviing completely
    # wrong value of gradients
    model.weights.grad.zero_()
    model.weights.grad.zero_()

    print(f'epoch:{epoch+1},loss is :{loss}')
    
     
```

**Model Evaluation**

```python
with torch.no_grad():
    y_pred=model.forward(x_test_tensor)
    print(y_pred)
```

```python
# As predicted values are between 0.1 and 0.9 and we have 0 and 1 only as output, so we change value of y_pred as 
# 1 if it is more than 0.8 and 0 if it is less than 0.5

y_pred=(y_pred>0.8).float()
accuracy=(y_pred==y_test_tensor).float().mean()
print(f'Accuracy:{accuracy}')
```
